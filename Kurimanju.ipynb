THE KURI MANJU PROBLEM (CHESTNUT BUNS)

```
## Import Liabry 
import math
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```
## Defind the function for the solar system
## Let say the solar size is about 2000000
def solar_system(object_size):
  solar_size = 2000000
  time = 0
  while True:
    if object_size>=solar_size:
      break
    else:
        object_size*=2
        time+=5
  return print("{} Minutes".format(time))
```

```
solar_system(300)
```

TOKYO DEMO

```
## Defind the function for the Tokyo Demo
## Let say the Tokyo Demo size is about 2870000
def Tokyo_Dome(object_size):
    size_tokyo_dome=2870000 
    time=0
    minutes=[]
    size=[]
    while True:
        if object_size>=size_tokyo_dome:
            break
        else:
            object_size*=2
            time+=5
            minutes.append(time)
            size.append(object_size)
    
    plt.plot(minutes, size, 'b-x', linewidth=4, markersize=10, markeredgewidth=5, markeredgecolor='red', linestyle='dashed')
    plt.xlabel("MINUTES"); 
    plt.ylabel("OBJECT");
    plt.title("TOKYO DEMO")
    
    return print("{} Minutes".format(time))
Tokyo_Dome(90000)
```
