```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.path as mpath
import matplotlib.patches as mpatches
%matplotlib inline
```

# PROBLEM 1

Choosing a dart mat

```python
def roll_dice():
    dice = np.random.choice(range(1, 7))
    return dice
```

# PROBLEM 2

Throw darts and record the place

```python
def throw_darts():
    mu = (roll_dice(),0)
    std = ((0.01, 0), (0, 0.1))
    
    x, y = np.random.multivariate_normal(mu, std)
    return x, y
print(throw_darts())
```

# PROBLEM 3

Throw a lot of darts and record them in a state of being tied to the number

```python
target = {1:[], 2:[], 3:[], 4:[], 5:[], 6:[]}
t_count = 1000

for _ in range(t_count):
    xy = throw_darts()
    t_no = round(xy[0])
    target[t_no].append(xy)
```

# PROBLEM 4

Throw a lot of darts and record them in a state of being tied to the number

```python
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot()

for t_no, xy in target.items():
    xy = np.array(xy)

    
    ax.scatter(xy[:, 0], xy[:, 1])

    
    inside_circle = mpatches.Circle((t_no,0),0.15, fc="none", ec="blue", alpha=1)
    midside_circle = mpatches.Circle((t_no,0),0.3, fc="none", ec="blue", alpha=1)
    outside_circle = mpatches.Circle((t_no,0),0.45, fc="none", ec="blue", alpha=1)
    ax.add_patch(inside_circle)
    ax.add_patch(midside_circle)
    ax.add_patch(outside_circle)

ax.set_aspect("equal")
```

# PROBLEM 5

 Functionalization

```python
def play_darts(num_of_target, throw_count):
    
    v = 0.1
    
   ## Createing a target for darts
    targets = {i:[] for i in range(1, num_of_target + 1)}
    
    ## Randomly throw darts
    for _ in range(throw_count):
        t_no = np.random.choice(range(1, num_of_target + 1))
        x = np.random.normal(t_no, v)
        y = np.random.normal(0, v)
        targets[t_no].append([x, y])
    
    ## Preparation of Graph
    fig = plt.figure(figsize=(20, 10))
    ax  = fig.add_subplot()
    
   ## Drawing darts result
    for t_no, xy in targets.items():
        xy = np.array(xy)
      ## Drawing the position of darts where they are stuck
        if len(xy) > 0:
            ax.scatter(xy[:, 0], xy[:, 1])

       ## Createing target
        inside_circle = mpatches.Circle((t_no,0),0.15, fc="none", ec="blue", alpha=1)
        midside_circle = mpatches.Circle((t_no,0),0.3, fc="none", ec="blue", alpha=1)
        outside_circle = mpatches.Circle((t_no,0),0.45, fc="none", ec="blue", alpha=1)
        ax.add_patch(inside_circle)
        ax.add_patch(midside_circle)
        ax.add_patch(outside_circle)        

    ax.set_aspect("equal")
    plt.show()
```

```python
play_darts(2,3)
play_darts(5,100)
play_darts(10,1000)
```
